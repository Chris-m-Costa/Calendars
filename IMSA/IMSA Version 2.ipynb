```python
driver = webdriver.Chrome()
driver.get("https://www.imsa.com/weathertech/weathertech-2024-schedule/")

# Wait for dynamic content to load if needed

sched_soup = BeautifulSoup(driver.page_source, 'html.parser')
schedule_list = sched_soup.find_all('div', {'class': 'schedule-item'})

for event in schedule_list:
    hero = event.find('h2', {'class': 'schedule-title'}).text
    
    link = event.find('a')
    if link:
        href = link.get('href').strip(' ')
        
    driver = webdriver.Chrome()
    driver.get(href)
 # Wait for dynamic content to load if needed
    event_soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    circuit = event_soup.find('div', {'class': 'track-name'})
  
    
    if circuit == 'Daytona International Speedway' or 'Sebring International Raceway' or 'Michelin Raceway Road Atlanta':
        event_list = event_soup.find('div', {'class': 'race-event-schedule-container'})
        for event_container in event_list.find_all('div', class_='race-event-schedule-container-inner'):
            for item in event_container.children:
                if item.name == 'div' and 'day-event-header' in item.get('class', []):
                    date = item.text.split()
                elif item.name == 'div' and 'day-event-details-container' in item.get('class', []):
                    time = item.find('div', class_='event-time').text.strip('ET').split('to')

                    start_time = time[0]
                    start_hour = start_time.split(':')[0]
                    if 'AM' in start_time:
                        if start_hour == '12':
                            start_hours.append(0)
                        else:   
                            start_hours.append(int(start_hour))

                    elif 'PM' in start_time:
                        start_hour = (int(start_hour) + 12)
                        if start_hour == 24:
                            start_hours.append(12)
                        else:
                            start_hours.append(start_hour)
                    start_minute = start_time.split(':')[1].strip(' AM').strip(' P')
        #             print(start_minute)
                    start_minutes.append(int(start_minute))

                    end_time = time[1]
                    end_hour = end_time.split(':')[0]
                    if 'AM' in end_time:
                        if end_hour == '12':
                            end_hours.append(0)
                        else:   
                            end_hours.append(int(end_hour))

                    elif 'PM' in end_time:
                        end_hour = (int(end_hour) + 12)
                        if end_hour == 24:
                            end_hours.append(12)
                        else:
                            end_hours.append(end_hour)

                    end_minute = end_time.split(':')[1].strip().strip(' AM').strip(' P')
        #             print(end_minute)
                    end_minutes.append(int(end_minute))

                    month = month_mapping.get(date[1])
                    months.append(month)
                    day = int(date[2].strip(','))
                    days.append(day)

                    desc = item.find('div', class_='event-name').text.strip()
                    descs.append(desc)

                    circuit = event_soup.find('div', {'class': 'track-name'}).text
                    circuits.append(circuit)
#                     hero = event_soup.find('div', {'class': 'hero-title'}).text.strip('\n')
                    titles.append(hero)    
    
    else:
        event_list = event_soup.find('div', {'class': 'race-event-schedule-container'})
        event_container = event_list.find('div', class_='race-event-schedule-container-inner')
        day_event_headers = event_container.find_all('div', class_='day-event-header')
        last_day_event_header = day_event_headers[-1]

        # Find the last 'day-event-details-container' after the last 'day-event-header'
        last_event_details_container = last_day_event_header.find_next_sibling('div', class_='day-event-details-container')[-1]

        # Find the last 'event-time' within the 'last_event_details_container'
        last_event_time = last_event_details_container.find('div', class_='event-time')[-1]

        time = last_event_time.text.strip('ET').split('to')

        start_time = time[0]
        start_hour = start_time.split(':')[0]
        if 'AM' in start_time:
            if start_hour == '12':
                start_hours.append(0)
            else:
                start_hours.append(int(start_hour))
        elif 'PM' in start_time:
            start_hour = (int(start_hour) + 12) % 24
            start_hours.append(start_hour)

        start_minute = start_time.split(':')[1].strip(' AM').strip(' P')
        start_minutes.append(int(start_minute))

        end_time = time[1]
        end_hour = end_time.split(':')[0]
        if 'AM' in end_time:
            if end_hour == '12':
                end_hours.append(0)
            else:
                end_hours.append(int(end_hour))
        elif 'PM' in end_time:
            end_hour = (int(end_hour) + 12) % 24
            end_hours.append(end_hour)

        end_minute = end_time.split(':')[1].strip().strip(' AM').strip(' P')
        end_minutes.append(int(end_minute))

        date = last_day_event_header.text.split()
        month = month_mapping.get(date[1])
        months.append(month)
        day = int(date[2].strip(','))
        days.append(day)

        desc = last_event_details_container.find('div', class_='event-name').text.strip()
        descs.append(desc)

        circuit = event_soup.find('div', {'class': 'track-name'}).text
        circuits.append(circuit)
        titles.append(hero)

# Quit the driver
driver.quit()

       
```
