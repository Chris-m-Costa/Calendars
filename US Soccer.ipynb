```python
driver = webdriver.Chrome()
driver.get("https://www.ussoccer.com/all-matches#upcoming-schedule")

# Wait for dynamic content to load if needed

sched_soup = BeautifulSoup(driver.page_source, 'html.parser')
m_schedule_details = sched_soup.find_all('div', {'class': 'MatchListItem-module__MatchListItem--3FWfJ MatchListItem-module__men--3m53_'})

driver.quit()

# <div class="MatchListItem-module__MatchListItem--3FWfJ MatchListItem-module__men--3m53_">
```

```python
days = [] 
months = []
titles = [] 
descs = []
stadiums = [] 
years = []

start_h = []
start_m = []
end_h = []
end_m = []


month_mapping = {
        'Jan': 1,
        'JAN': 1,
        'January': 1,
        'Feb': 2,
        'FEB': 2,
        'Mar': 3,
        'March': 3,
        'Apr': 4,
        'May': 5,
        'Jun': 6,
        'Jul': 7,
        'Aug': 8,
        'Sep': 9,
        'Oct': 10,
        'October': 10,
        'Nov': 11,
        'Dec': 12,
        'DEC': 12
    }

```

```python

for i in w_schedule_details:
    if 'USWNT' in i.find_all('div', {'class':"MatchListItem-module__team--1WSAM"})[0].text or 'USWNT' in i.find_all('div', {'class':"MatchListItem-module__team--1WSAM"})[1].text:
    
        temp_stadium = i.find('div', {'class': 'MatchListItem-module__venue--75PXp'}).text
        stadiums.append(temp_stadium)

    #   <div class="MatchListItem-module__dateTime--2qJRC">Feb 06 2024 - 1:00PM EST</div>

        temp_datetime = i.find('div', {'class':"MatchListItem-module__dateTime--2qJRC"}).text
        temp_date = temp_datetime.split(' - ')[0]

        temp_day = int(temp_date.split(' ')[1].lstrip('0'))
        days.append(temp_day)

        temp_month = temp_date.split(' ')[0].lstrip('0')
        converted_month = month_mapping[temp_month]
        months.append(converted_month)

        year = int(temp_date.split(' ')[2])
        years.append(year)

        temp_time = temp_datetime.split(' - ')[1]
        if temp_time == 'TBD':
            start_h.append(18)
            start_m.append(0)
            end_h.append(20)
            end_m.append(0)
        else:    
            temp_start_h = int(temp_time.split(':')[0])
            if 'PM' in temp_time:
                if temp_start_h == 12:
                    start_h.append(temp_start_h)
                else:
                    temp_start_h += 12
                    start_h.append(temp_start_h)
            else:
                start_h.append(temp_start_h)

            temp_end_h = temp_start_h +2
            if temp_end_h == 24:
                end_h.append(0)
            else:
                end_h.append(temp_end_h)



    #         temp_m = temp_time.split(':')[1]
    #         if 'EST' in temp_m:
    #             temp_m = temp_m[:-6]
    #         elif 'EDT' in temp_m:
    #             temp_m = temp_m[:-6]
    #         start_m.append(int(temp_m))
    #         end_m.append(int(temp_m))

            temp_m = temp_time.split(':')[1]
            temp_m = temp_m[:-6]
            start_m.append(int(temp_m))
            end_m.append(int(temp_m))

        # <div class="MatchListItem-module__team--1WSAM">U-17 WNT</div>
        # <div class="MatchListItem-module__vs--3VSc_">VS</div>
        # <div class="MatchListItem-module__team--1WSAM">CAN</div>

        tit1 = i.find_all('div', {'class':"MatchListItem-module__team--1WSAM"})[0].text
        tit2 = i.find('div', {'class':"MatchListItem-module__vs--3VSc_"}).text
        tit3 = i.find_all('div', {'class':"MatchListItem-module__team--1WSAM"})[1].text
        temp_title = tit1 + ' ' + tit2 + ' ' + tit3
        titles.append(temp_title)

    #     <div class="MatchListItem-module__title--eQgH3">2024 Concacaf Women's U-17 Championship</div>
        temp_desc = i.find('div', {'class':"MatchListItem-module__title--eQgH3"}).text
        descs.append(temp_desc)
        
    else:
        pass
```

```python
import pytz
import icalendar
from datetime import datetime
from pathlib import Path
import os
from icalendar import Calendar, Event,  vText, vCalAddress
```

```python
US_Soccer = Calendar()
US_Soccer.add('prodid', '-//My calendar product//example.com//')
US_Soccer.add('version', '2.0')
uid = 110
```

```python
 
for title, stadium, day, month, desc, s_h, s_m, e_h, e_m, year in zip(titles, stadiums, days, months, descs, start_h, start_m, end_h, end_m, years):
    ievent = Event()
    ievent.add('summary', title) #Title of the event
    ievent.add('description', desc)     #Description of event
    #     ievent.add('description', f"{event}\n{length}")

    ievent.add('dtstart', datetime(year, month, day, s_h, s_m, 0, tzinfo=pytz.timezone('est')))
    if e_h == 0:
        ievent.add('dtend', datetime(year, month, day+1, e_h, e_m, 0, tzinfo=pytz.timezone('est')))
    else:    
        ievent.add('dtend', datetime(year, month, day, e_h, e_m, 0, tzinfo=pytz.timezone('est')))
    ievent.add('dtstamp', datetime(2024, 1, 16, 0, 10, 0, tzinfo=pytz.timezone('est')))
    ievent.add('location', stadium)
    ievent.add('uid', uid)
    uid += 1
    US_Soccer.add_component(ievent)
```

```python
directory = str(Path(r'C:\Users\chris\Documents\Calendars')) + "/"
print("ics file will be generated at ", directory)
f = open(os.path.join(directory, 'US Soccer.ics'), 'wb')
f.write(US_Soccer.to_ical())
f.close()
```
