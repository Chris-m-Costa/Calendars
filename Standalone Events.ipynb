```python
import pytz
import icalendar
from datetime import datetime
from pathlib import Path
import os
from icalendar import Calendar, Event,  vText, vCalAddress
```

```python
Standalone_Events = Calendar()
Standalone_Events.add('prodid', '-//My calendar product//example.com//')
Standalone_Events.add('version', '2.0')
```

```python


uid = 210 
for title, circuit, day, e_day, month, desc, s_h, s_m, e_h, e_m, tz in zip(titles, circuits, days, end_day, months, descs, start_h, start_m, end_h, end_m, timezones):
    ievent = Event()
    ievent.add('summary', title) #Title of the event
    ievent.add('description', desc)     #Description of event
    #     ievent.add('description', f"{event}\n{length}")

    ievent.add('dtstart', datetime(year, month, day, s_h, s_m, 0, tzinfo=pytz.timezone(tz)))
    ievent.add('dtend', datetime(year, month, e_day, e_h, e_m, 0, tzinfo=pytz.timezone(tz)))
    ievent.add('dtstamp', datetime(2024, 1, 16, 0, 10, 0, tzinfo=pytz.timezone('est')))
    ievent.add('location', circuit)
    ievent.add('uid', uid)
    uid += 1
    Standalone_Events.add_component(ievent)
```
